#1. An√°lisis de Facturaci√≥n
* Ventas Totales y N√∫mero de Transacciones. Facturaci√≥n total de cada tienda, suma de ingreso total consolidado y anual por tienda. Cu√°l tienda factura menos ?


```python
#print(f'\nFacturaci√≥n total de cada tienda en Lista: ')

# DataFrames: tiendas

facturacion_tienda = [float(tienda['Precio'].sum()),float(tienda2['Precio'].sum()),float(tienda3['Precio'].sum()),float(tienda4['Precio'].sum())]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Encontrar el valor m√≠nimo de facturaci√≥n
facturacion_minima = min(facturacion_tienda)


# Encontrar el √≠ndice del valor m√≠nimo
facturacion_minima_indice = facturacion_tienda.index(facturacion_minima)

# Imprimir el resultado
print(f'\n-------------------------------------------------------------------------------------')
print(f'\n**La tienda que tiene menor facturaci√≥n es la tienda # {facturacion_minima_indice + 1} con un valor de ${facturacion_minima:,.2f}**')

# Visualizaci√≥n facturaci√≥n de cada tienda
print(f'\n-------------------------------------------------------------------------------------')
print(f'\nFacturaci√≥n total de cada tienda: ')
print(f'\n')

for i in range(0,len(facturacion_tienda)):
  print(f"\t- Tienda {i+1} es: ${facturacion_tienda[i]:,.2f}")
print(f'\n-------------------------------------------------------------------------------------')
```

##üìà Gr√°fico 1 (l√≠neas) : Comparaci√≥n de Ventas Totales por Tienda


```python
import matplotlib.pyplot as plt
```

```python
# Datos de ingresos totales por tienda
# Lista facturacion_tienda generada item anterior

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear el gr√°fico de lineas
plt.figure(figsize=(10, 6),dpi = 110)
plt.plot(tiendas, facturacion_tienda, color ='b', linestyle = '--', lw = 3, marker = 'o')

# Personalizar el gr√°fico
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos Totales')
plt.grid(True)
plt.legend();
```

##üìà Gr√°fico 2 (l√≠neas) : Comparaci√≥n de Ventas Totales por a√±o por Tienda

* Facturacion por tienda a lo largo de los a√±os

```python
import matplotlib.pyplot as plt
import pandas as pd

# DataFrames: tiendas
tiendas_bd = [tienda, tienda2, tienda3, tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Concatenar todos los DataFrames en uno solo
all_data = pd.concat(tiendas_bd, keys=tiendas)
all_data = all_data.reset_index(level=0)
all_data = all_data.rename(columns={'level_0': 'Tienda'}) # Renombrar la columna 'level_0' a 'Tienda'

# Extraer el a√±o de la columna 'Fecha de Compra'
all_data['A√±o'] = pd.to_datetime(all_data['Fecha de Compra']).dt.year

# Agrupar por tienda y a√±o, sumando las ventas
ventas_por_tienda_a√±o = all_data.groupby(['Tienda', 'A√±o'])['Precio'].sum().reset_index()

# Crear el gr√°fico de l√≠neas
fig, ax = plt.subplots(figsize=(10, 6))

for tienda in tiendas:
    data_tienda = ventas_por_tienda_a√±o[ventas_por_tienda_a√±o['Tienda'] == tienda]
    ax.plot(data_tienda['A√±o'], data_tienda['Precio'], label=tienda)

# Personalizar el gr√°fico
ax.set_xlabel('A√±o')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por tienda a lo largo de los a√±os')
ax.legend()
ax.grid(True)

# Mostrar el gr√°fico
plt.show()
```

```python
# Configurar el formato de visualizaci√≥n de los n√∫meros float
pd.options.display.float_format = '{:.2f}'.format

# Crear tabla resumen con valores por a√±o y total
tabla_resumen = ventas_por_tienda_a√±o.pivot(index='Tienda', columns='A√±o', values='Precio').fillna(0)  # Llenar valores NaN con 0
tabla_resumen['Total'] = tabla_resumen.sum(axis=1)  # Agregar columna Total
tabla_resumen= tabla_resumen.astype(float)  # Convertir a tipo float
print("\nTabla Resumen:")
print(tabla_resumen)
```

**Analisis** : Aunque la Tienda 3 sigue teniendo un rendimiento menor en varios aspectos, la Tienda 4 es la que ha experimentado la mayor ca√≠da en las ventas en los √∫ltimos a√±os, como se puede apreciar en la gr√°fica.

Por lo tanto, ser√≠a importante revisar con m√°s detalle la situaci√≥n de la Tienda 4 y considerar si esta ca√≠da en las ventas es un factor determinante en la decisi√≥n de cierre,

# 2. Ventas por categor√≠a

Evidenciar las categor√≠as m√°s vendidas en cada tienda y analizar cantidad de ventas por cada tienda agrupadas por categor√≠a.

```python
#tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
#tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener la lista de categor√≠as no repetidas
categorias = set()
for tienda_df in tiendas_bd:  # Usar tiendas_bd en lugar de tiendas
    categorias.update(tienda_df['Categor√≠a del Producto'].unique())
categorias = sorted(list(categorias))  # Ordenar alfab√©ticamente

# Crear un diccionario para almacenar los datos
datos_por_categoria = {}
for categoria in categorias:
    datos_por_categoria[categoria] = {}
    for i, tienda_df in enumerate(tiendas_bd):  # Usar tiendas_bd en lugar de tiendas
        # Filtrar por categor√≠a y calcular ventas y cantidad de vendidas
        ventas_categoria = tienda_df[tienda_df['Categor√≠a del Producto'] == categoria]['Precio'].sum()
        cantidad_vendida = tienda_df[tienda_df['Categor√≠a del Producto'] == categoria]['Producto'].count()
        datos_por_categoria[categoria][tiendas[i]] = {'Ventas': ventas_categoria, 'Cantidad Vendida': cantidad_vendida}  # Usar tiendas para los nombres de las tiendas
```

```python
from prettytable import PrettyTable

#tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
#tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear la tabla con PrettyTable
tabla = PrettyTable()
tabla.field_names = ["Categor√≠a"] + [f"{tienda} {tipo}" for tienda in tiendas for tipo in  ["Ventas", "Cantidad Vendida"]]

# Agregar filas a la tabla con estilo de l√≠neas alternadas
for i, (categoria, datos_tiendas) in enumerate(datos_por_categoria.items()):
    fila = [categoria]
    for tienda in tiendas:
        fila.extend([datos_tiendas[tienda]['Ventas'], datos_tiendas[tienda]['Cantidad Vendida']])
    tabla.add_row(fila)


tabla.add_row(["", "", "", "", "", "", "", "", ""])  # Agregar una fila vac√≠a para el espacio gris (opcional)

# Agregar fila TOTAL
fila_total = ["TOTAL"]
for tienda in tiendas:
    total_ventas = sum(datos_tiendas[tienda]['Ventas'] for datos_tiendas in datos_por_categoria.values())
    total_cantidad_vendida = sum(datos_tiendas[tienda]['Cantidad Vendida'] for datos_tiendas in datos_por_categoria.values())
    fila_total.extend([total_ventas, total_cantidad_vendida])
tabla.add_row(fila_total)

# Imprimir la tabla
print(tabla)
```

##üìà Gr√°fico 1 (Barras): Ventas por Categor√≠a por Tienda

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para el gr√°fico
categorias = list(datos_por_categoria.keys())
num_tiendas = len(tiendas)
ancho_barra = 0.8 / num_tiendas  # Ajustar ancho de barras seg√∫n la cantidad de tiendas

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(12, 6))

# Crear barras para cada tienda
for i, tienda in enumerate(tiendas):
    ventas = [datos_por_categoria[categoria][tienda]['Ventas'] for categoria in categorias]
    x_pos = np.arange(len(categorias)) + i * ancho_barra  # Ajustar posici√≥n de barras
    ax.bar(x_pos, ventas, width=ancho_barra, label=tienda)

# Personalizar el gr√°fico
ax.set_xlabel('Categor√≠a')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por Categor√≠a por Tienda')
ax.set_xticks(np.arange(len(categorias)) + ancho_barra * (num_tiendas - 1) / 2)  # Centrar etiquetas de categor√≠as
ax.set_xticklabels(categorias, rotation=45, ha='right')  # Rotar etiquetas de categor√≠as si es necesario
ax.legend()
plt.tight_layout()  # Ajustar espacio entre elementos del gr√°fico
plt.show()
```

* ***Analisis :***
-	Las 3 categor√≠as que generan la mayor facturaci√≥n en las 4 tiendan son Electr√≥nicos 18,8%, Electrodom√©sticos 12,2% y Muebles 20% en las cuatro tiendas. Liderando en facturaci√≥n la Tienda 1 en dos de las categor√≠as.
-	Las 3 categor√≠as que generan la menor facturaci√≥n son Libros, art√≠culos para el Hogar y Juguetes.
-	La tienda 1 es la que menos ventas tiene en libros y art√≠culos para el hogar.
-	La tienda 2 es la que menos ventas tiene en juguetes.

# 3. Calificaci√≥n promedio de la tienda
## Analizar el promedio de evaluaci√≥n de los clientes para cada tienda, como indicador de satisfacci√≥n del cliente.

```python
tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, tienda_df in enumerate(tiendas_bd):
    # Aseg√∫rate de que los datos de la columna "Calificaci√≥n" sean de tipo num√©rico
    tienda_df["Calificaci√≥n"] = pd.to_numeric(tienda_df["Calificaci√≥n"], errors='coerce')

    # Calcular la calificaci√≥n promedio manualmente
    suma_calificaciones = 0
    num_calificaciones = 0
    for calificacion in tienda_df["Calificaci√≥n"]:
        if not pd.isna(calificacion):  # Ignorar valores NaN
            suma_calificaciones += calificacion
            num_calificaciones += 1

    calificacion_promedio = suma_calificaciones / num_calificaciones if num_calificaciones else 0  # Evitar divisi√≥n por cero

    # Imprimir el resultado
    print(f"Tienda {i + 1}: Calificaci√≥n promedio = {calificacion_promedio:.2f}")
```

##üìà Gr√°fico 1 (Barras): Calificaci√≥n promedio por Tienda consolidado

```python
import matplotlib.pyplot as plt

tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

calificaciones_promedio = []
for tienda_df in tiendas_bd:
    calificaciones_promedio.append(tienda_df["Calificaci√≥n"].mean())

# Crear el gr√°fico de barras
plt.bar(tiendas, calificaciones_promedio, color=['blue', 'green', 'red', 'orange'])

# Personalizar el gr√°fico
plt.title('Calificaci√≥n Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificaci√≥n Promedio')
plt.ylim(3.0, 4.5)  # Ajustar el rango del eje y para mejor visualizaci√≥n
plt.grid(True)

# Mostrar el gr√°fico
plt.show()
```

# 4. Productos m√°s y menos vendidos en cada tienda

Tabla 1: Producto m√°s vendido y menos vendido por tienda

```python
# DataFrames: tiendas
#tiendas_bd = [tienda, tienda2, tienda3, tienda4]
#tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, tienda_df in enumerate(tiendas_bd):
    # Agrupar por producto y contar las ventas
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')

    # Ordenar por cantidad vendida (descendente para el m√°s vendido, ascendente para el menos vendido)
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]

    # Obtener la categor√≠a del producto m√°s vendido y menos vendido
    categoria_mas_vendido = tienda_df[tienda_df['Producto'] == mas_vendido['Producto']]['Categor√≠a del Producto'].iloc[0]
    categoria_menos_vendido = tienda_df[tienda_df['Producto'] == menos_vendido['Producto']]['Categor√≠a del Producto'].iloc[0]

    # Calcular la calificaci√≥n promedio manualmente
    suma_calificaciones = 0
    num_calificaciones = 0
    for calificacion in tienda_df["Calificaci√≥n"]:
        if not pd.isna(calificacion):  # Ignorar valores NaN
            suma_calificaciones += calificacion
            num_calificaciones += 1

    calificacion_promedio = suma_calificaciones / num_calificaciones if num_calificaciones else 0  # Evitar divisi√≥n por cero

    # Imprimir los resultados
    print(f"Tienda {i + 1}:")
    print(f"  Producto m√°s vendido: {mas_vendido['Producto']} ({mas_vendido['Cantidad Vendida']} ventas) - Categor√≠a: {categoria_mas_vendido}")
    print(f"  Producto menos vendido: {menos_vendido['Producto']} ({menos_vendido['Cantidad Vendida']} ventas) - Categor√≠a: {categoria_menos_vendido}")
    print(f"  Calificaci√≥n promedio: {calificacion_promedio:.2f}")  # Imprimir la calificaci√≥n promedio
    print("-" * 20)  # Separador entre tiendas
```
