#1. Análisis de Facturación
* Ventas Totales y Número de Transacciones. Facturación total de cada tienda, suma de ingreso total consolidado y anual por tienda. Cuál tienda factura menos ?


```python
#print(f'\nFacturación total de cada tienda en Lista: ')

# DataFrames: tiendas

facturacion_tienda = [float(tienda['Precio'].sum()),float(tienda2['Precio'].sum()),float(tienda3['Precio'].sum()),float(tienda4['Precio'].sum())]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Encontrar el valor mínimo de facturación
facturacion_minima = min(facturacion_tienda)


# Encontrar el índice del valor mínimo
facturacion_minima_indice = facturacion_tienda.index(facturacion_minima)

# Imprimir el resultado
print(f'\n-------------------------------------------------------------------------------------')
print(f'\n**La tienda que tiene menor facturación es la tienda # {facturacion_minima_indice + 1} con un valor de ${facturacion_minima:,.2f}**')

# Visualización facturación de cada tienda
print(f'\n-------------------------------------------------------------------------------------')
print(f'\nFacturación total de cada tienda: ')
print(f'\n')

for i in range(0,len(facturacion_tienda)):
  print(f"\t- Tienda {i+1} es: ${facturacion_tienda[i]:,.2f}")
print(f'\n-------------------------------------------------------------------------------------')
```

##📈 Gráfico 2 (líneas) : Comparación de Ventas Totales por año por Tienda

* Facturacion por tienda a lo largo de los años

```python
import matplotlib.pyplot as plt
import pandas as pd

# DataFrames: tiendas
tiendas_bd = [tienda, tienda2, tienda3, tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Concatenar todos los DataFrames en uno solo
all_data = pd.concat(tiendas_bd, keys=tiendas)
all_data = all_data.reset_index(level=0)
all_data = all_data.rename(columns={'level_0': 'Tienda'}) # Renombrar la columna 'level_0' a 'Tienda'

# Extraer el año de la columna 'Fecha de Compra'
all_data['Año'] = pd.to_datetime(all_data['Fecha de Compra']).dt.year

# Agrupar por tienda y año, sumando las ventas
ventas_por_tienda_año = all_data.groupby(['Tienda', 'Año'])['Precio'].sum().reset_index()

# Crear el gráfico de líneas
fig, ax = plt.subplots(figsize=(10, 6))

for tienda in tiendas:
    data_tienda = ventas_por_tienda_año[ventas_por_tienda_año['Tienda'] == tienda]
    ax.plot(data_tienda['Año'], data_tienda['Precio'], label=tienda)

# Personalizar el gráfico
ax.set_xlabel('Año')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por tienda a lo largo de los años')
ax.legend()
ax.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Configurar el formato de visualización de los números float
pd.options.display.float_format = '{:.2f}'.format

# Crear tabla resumen con valores por año y total
tabla_resumen = ventas_por_tienda_año.pivot(index='Tienda', columns='Año', values='Precio').fillna(0)  # Llenar valores NaN con 0
tabla_resumen['Total'] = tabla_resumen.sum(axis=1)  # Agregar columna Total
tabla_resumen= tabla_resumen.astype(float)  # Convertir a tipo float
print("\nTabla Resumen:")
print(tabla_resumen)
```

**Analisis** : Aunque la Tienda 3 sigue teniendo un rendimiento menor en varios aspectos, la Tienda 4 es la que ha experimentado la mayor caída en las ventas en los últimos años, como se puede apreciar en la gráfica.

Por lo tanto, sería importante revisar con más detalle la situación de la Tienda 4 y considerar si esta caída en las ventas es un factor determinante en la decisión de cierre,

# 2. Ventas por categoría

Evidenciar las categorías más vendidas en cada tienda y analizar cantidad de ventas por cada tienda agrupadas por categoría.

```python
#tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
#tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener la lista de categorías no repetidas
categorias = set()
for tienda_df in tiendas_bd:  # Usar tiendas_bd en lugar de tiendas
    categorias.update(tienda_df['Categoría del Producto'].unique())
categorias = sorted(list(categorias))  # Ordenar alfabéticamente

# Crear un diccionario para almacenar los datos
datos_por_categoria = {}
for categoria in categorias:
    datos_por_categoria[categoria] = {}
    for i, tienda_df in enumerate(tiendas_bd):  # Usar tiendas_bd en lugar de tiendas
        # Filtrar por categoría y calcular ventas y cantidad de vendidas
        ventas_categoria = tienda_df[tienda_df['Categoría del Producto'] == categoria]['Precio'].sum()
        cantidad_vendida = tienda_df[tienda_df['Categoría del Producto'] == categoria]['Producto'].count()
        datos_por_categoria[categoria][tiendas[i]] = {'Ventas': ventas_categoria, 'Cantidad Vendida': cantidad_vendida}  # Usar tiendas para los nombres de las tiendas
```

```python
from prettytable import PrettyTable

#tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
#tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear la tabla con PrettyTable
tabla = PrettyTable()
tabla.field_names = ["Categoría"] + [f"{tienda} {tipo}" for tienda in tiendas for tipo in  ["Ventas", "Cantidad Vendida"]]

# Agregar filas a la tabla con estilo de líneas alternadas
for i, (categoria, datos_tiendas) in enumerate(datos_por_categoria.items()):
    fila = [categoria]
    for tienda in tiendas:
        fila.extend([datos_tiendas[tienda]['Ventas'], datos_tiendas[tienda]['Cantidad Vendida']])
    tabla.add_row(fila)


tabla.add_row(["", "", "", "", "", "", "", "", ""])  # Agregar una fila vacía para el espacio gris (opcional)

# Agregar fila TOTAL
fila_total = ["TOTAL"]
for tienda in tiendas:
    total_ventas = sum(datos_tiendas[tienda]['Ventas'] for datos_tiendas in datos_por_categoria.values())
    total_cantidad_vendida = sum(datos_tiendas[tienda]['Cantidad Vendida'] for datos_tiendas in datos_por_categoria.values())
    fila_total.extend([total_ventas, total_cantidad_vendida])
tabla.add_row(fila_total)

# Imprimir la tabla
print(tabla)
```

##📈 Gráfico 1 (Barras): Ventas por Categoría por Tienda

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para el gráfico
categorias = list(datos_por_categoria.keys())
num_tiendas = len(tiendas)
ancho_barra = 0.8 / num_tiendas  # Ajustar ancho de barras según la cantidad de tiendas

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(12, 6))

# Crear barras para cada tienda
for i, tienda in enumerate(tiendas):
    ventas = [datos_por_categoria[categoria][tienda]['Ventas'] for categoria in categorias]
    x_pos = np.arange(len(categorias)) + i * ancho_barra  # Ajustar posición de barras
    ax.bar(x_pos, ventas, width=ancho_barra, label=tienda)

# Personalizar el gráfico
ax.set_xlabel('Categoría')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por Categoría por Tienda')
ax.set_xticks(np.arange(len(categorias)) + ancho_barra * (num_tiendas - 1) / 2)  # Centrar etiquetas de categorías
ax.set_xticklabels(categorias, rotation=45, ha='right')  # Rotar etiquetas de categorías si es necesario
ax.legend()
plt.tight_layout()  # Ajustar espacio entre elementos del gráfico
plt.show()
```

* ***Analisis :***
-	Las 3 categorías que generan la mayor facturación en las 4 tiendan son Electrónicos 18,8%, Electrodomésticos 12,2% y Muebles 20% en las cuatro tiendas. Liderando en facturación la Tienda 1 en dos de las categorías.
-	Las 3 categorías que generan la menor facturación son Libros, artículos para el Hogar y Juguetes.
-	La tienda 1 es la que menos ventas tiene en libros y artículos para el hogar.
-	La tienda 2 es la que menos ventas tiene en juguetes.
-	La tienda 4, presenta el menor volumen de ventas en la mayoría de las categorías. seguida por la tienda 3

# 3. Calificación promedio de la tienda
## Analizar el promedio de evaluación de los clientes para cada tienda, como indicador de satisfacción del cliente.

```python
tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, tienda_df in enumerate(tiendas_bd):
    # Asegúrate de que los datos de la columna "Calificación" sean de tipo numérico
    tienda_df["Calificación"] = pd.to_numeric(tienda_df["Calificación"], errors='coerce')

    # Calcular la calificación promedio manualmente
    suma_calificaciones = 0
    num_calificaciones = 0
    for calificacion in tienda_df["Calificación"]:
        if not pd.isna(calificacion):  # Ignorar valores NaN
            suma_calificaciones += calificacion
            num_calificaciones += 1

    calificacion_promedio = suma_calificaciones / num_calificaciones if num_calificaciones else 0  # Evitar división por cero

    # Imprimir el resultado
    print(f"Tienda {i + 1}: Calificación promedio = {calificacion_promedio:.2f}")
```

# 4. Productos más y menos vendidos en cada tienda

Tabla 1: Producto más vendido y menos vendido por tienda

```python
# DataFrames: tiendas
#tiendas_bd = [tienda, tienda2, tienda3, tienda4]
#tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, tienda_df in enumerate(tiendas_bd):
    # Agrupar por producto y contar las ventas
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')

    # Ordenar por cantidad vendida (descendente para el más vendido, ascendente para el menos vendido)
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]

    # Obtener la categoría del producto más vendido y menos vendido
    categoria_mas_vendido = tienda_df[tienda_df['Producto'] == mas_vendido['Producto']]['Categoría del Producto'].iloc[0]
    categoria_menos_vendido = tienda_df[tienda_df['Producto'] == menos_vendido['Producto']]['Categoría del Producto'].iloc[0]

    # Calcular la calificación promedio manualmente
    suma_calificaciones = 0
    num_calificaciones = 0
    for calificacion in tienda_df["Calificación"]:
        if not pd.isna(calificacion):  # Ignorar valores NaN
            suma_calificaciones += calificacion
            num_calificaciones += 1

    calificacion_promedio = suma_calificaciones / num_calificaciones if num_calificaciones else 0  # Evitar división por cero

    # Imprimir los resultados
    print(f"Tienda {i + 1}:")
    print(f"  Producto más vendido: {mas_vendido['Producto']} ({mas_vendido['Cantidad Vendida']} ventas) - Categoría: {categoria_mas_vendido}")
    print(f"  Producto menos vendido: {menos_vendido['Producto']} ({menos_vendido['Cantidad Vendida']} ventas) - Categoría: {categoria_menos_vendido}")
    print(f"  Calificación promedio: {calificacion_promedio:.2f}")  # Imprimir la calificación promedio
    print("-" * 20)  # Separador entre tiendas
```

 - Al analizar esta información nos damos cuenta que en la Tienda 2 los productos mas vendidos estan en la categoria de Libros y Juguetes, que son categorias que no estan generando gran facturacion en valores.
 - Si solo validamos este item podriamos pensar en que la tienda 2 es la tienda menos rentable, y opcionada a ser cerrada.
